```python
import pandas as pd
import sqlalchemy as sa

# Verbinding maken met de XAMPP database
engine = sa.create_engine("mysql+pymysql://root:@localhost/climate_watch")

# Laden van de data (ByCountry als startpunt)
df = pd.read_sql("SELECT * FROM global_temp_country", engine)
df.head()
```

```python
# Checken per kolom waarden die missen
print("Missende waarden per kolom:")
print(df.isnull().sum())

# Checken per rij waarden die missen (toon de eerste 5 rijen met missende data)
print("\nRijen met missende waarden:")
print(df[df.isnull().any(axis=1)].head())
```
